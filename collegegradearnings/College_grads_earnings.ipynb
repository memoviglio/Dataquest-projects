```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads.tail()
```

```python
recent_grads = recent_grads.dropna()
```

```python
recent_grads.plot("Sample_size","Median",kind="scatter",title="median earnings vs. sample size")
```

```python
recent_grads.plot("Sample_size","Unemployment_rate",kind="scatter",title="sample size vs. unemployment",xlim=(0,3000))
```

```python
recent_grads.plot("Full_time","Median",kind="scatter",title="Median earnings vs. number of full-time professionals")
```

```python
recent_grads.plot("ShareWomen","Unemployment_rate",kind="scatter",title="Unemployment rates vs. female shares of majors")
```

```python
recent_grads.plot("Men","Median",kind="scatter",title="Median earnings vs. number of male graduates")
```

```python
recent_grads.plot("Women","Median",kind="scatter",title="Median earnings vs. number of female graduates")
```
