```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv",encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
yes_no = {"Yes":True, "No":False}
for col in star_wars.columns[1:3]:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
rename_column ={star_wars.columns[3]:"seen_1",
                star_wars.columns[4]:"seen_2",
                star_wars.columns[5]:"seen_3",
                star_wars.columns[6]:"seen_4",
                star_wars.columns[7]:"seen_5",
                star_wars.columns[8]:"seen_6"}

star_wars = star_wars.rename(columns=rename_column)
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
rank_column_rename = {star_wars.columns[9]:"ranking_1",
                      star_wars.columns[10]:"ranking_2",
                      star_wars.columns[11]:"ranking_3",
                      star_wars.columns[12]:"ranking_4",
                      star_wars.columns[13]:"ranking_5",
                      star_wars.columns[14]:"ranking_6"}
star_wars = star_wars.rename(columns = rank_column_rename)
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

So far, we have cleaned up null values and ambiguous column names, simplified responses to binary ones and revealed the liking rankings.It looks like the first 3 movies are more favorable by people, with episode V picked as most favorite by most people.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

First 3 movies are seen by more people as expected and this may explain why they are more favorable than last 3 movies. Only episode I is seen as much as the first 3 movies and this is why it ranks in favorite rankings.

```python
import matplotlib.pyplot as plt
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
print(males[star_wars.columns[9:15]].mean())
print(females[star_wars.columns[9:15]].mean())

males[star_wars.columns[9:15]].mean().plot.bar()
plt.show()
females[star_wars.columns[9:15]].mean().plot.bar()
plt.show()
```

While women like episode I much more than men, men like episode IV much more than women.

```python
males[star_wars.columns[3:9]].sum().plot.bar()
plt.show()
females[star_wars.columns[3:9]].sum().plot.bar()
plt.show()
```

```python
fan = star_wars[star_wars[star_wars.columns[2]] == True]
no_fan = star_wars[star_wars[star_wars.columns[2]] == False]
fan[star_wars.columns[9:15]].mean().plot.bar()
plt.show()
no_fan[star_wars.columns[9:15]].mean().plot.bar()
plt.show()
fan[star_wars.columns[3:9]].sum().plot.bar()
plt.show()
no_fan[star_wars.columns[3:9]].sum().plot.bar()
plt.show()
```

People who don't consider themselves as a fan favors "Episode I" as much as "Episode IV"
